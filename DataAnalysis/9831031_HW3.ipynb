Replication به فرآیند کپی کردن بلوک های داده از یک DataNode (گره ای در خوشه Hadoop که داده ها را ذخیره می کند) به چندین DataNode دیگر به منظور افزایش دوام و در دسترس بودن داده ها اشاره دارد.

هنگامی که یک فایل در یک سیستم فایل توزیع شده Hadoop (HDFS) ذخیره می‌شود، به بلوک‌ها تقسیم می‌شود و هر بلوک تعداد دفعات مشخصی در DataNode‌های مختلف در خوشه تکرار می‌شود. ضریب تکرار، که توسط مدیر تنظیم می شود، تعیین می کند که از هر بلوک چند نسخه باید ساخته شود. به عنوان مثال، ضریب تکرار 3 به این معنی است که هر بلوک دارای 3 نسخه ذخیره شده در DataNode های مختلف است.

Replication به محافظت در برابر از دست دادن داده ها در صورت خرابی گره کمک می کند و همچنین عملکرد دسترسی به داده ها را با اجازه دادن به چندین نسخه از بلوک های داده در دسترس در گره های مختلف بهبود می بخشد.

در HDFS (Hadoop Distributed File System)، یک بلوک کوچکترین واحد داده ای است که می تواند ذخیره و تکثیر شود. هنگامی که یک فایل در HDFS ذخیره می شود، به یک یا چند بلوک تقسیم می شود و سپس هر بلوک در یک DataNode (گره ای در خوشه Hadoop که داده ها را ذخیره می کند) ذخیره می شود. اندازه بلوک قابل تنظیم است و معمولاً روی مقدار زیادی مانند 128 مگابایت یا 256 مگابایت تنظیم می شود تا فضای ذخیره سازی و کارایی شبکه بهینه شود. اگر اندازه بلوک خیلی کوچک تنظیم شود، می تواند تأثیر منفی بر عملکرد و کارایی ذخیره سازی داشته باشد. از خوشه HDFS. به عنوان مثال، اندازه بلوک کوچک منجر به تعداد زیادی بلوک برای یک فایل می شود که می تواند هزینه های ذخیره سازی، مدیریت و تکرار بلوک ها را افزایش دهد. علاوه بر این، اندازه بلوک کوچک می‌تواند تعداد درخواست‌های شبکه را افزایش داده و سرعت انتقال داده را کاهش دهد، زیرا برای خواندن یا نوشتن یک فایل به درخواست‌های بیشتری نیاز است. از سوی دیگر، اگر اندازه بلوک خیلی بزرگ باشد، می‌تواند داده‌ها را ایجاد کند. دسترسی آهسته است زیرا زمان جستجو برای بلوک را افزایش می دهد. به طور کلی، انتخاب اندازه بلوک مناسب بر اساس نیازها و ویژگی های خاص داده ها و محیط خوشه مهم است.

<div align="left">
    
- Disk Based Frameworks 
    
            Persists intermediate results to disk 
    
            Data is reloaded from disk with every query  
    
            Easy failure recovery
    
            Best for ETL like work-loads 
    
            Examples: Hadoop, Dryad
    
- Memory Based Frameworks 
    
            Circumvents heavy cost of I/O by keeping intermediate results in memory 
    
            Sensitive to availability of memory 
    
            Remembers operations applied to dataset 
    
            Best for iterative workloads 
            
            Examples: Spark, Flink

</div>

#  اسپارک Shuffle 

    Spark SQL shuffle مکانیزمی است برای توزیع مجدد یا پارتیشن بندی مجدد داده ها به طوری که داده ها به طور متفاوت در پارتیشن ها گروه بندی می شوند، بر اساس اندازه داده شما ممکن است نیاز باشد تعداد پارتیشن های RDD/DataFrame را با استفاده از اسپارک کاهش یا افزایش دهید.
    برای مثال وقتی روی دو dataframe مختلف که روی شبکه توزیع شده اند
    دستور join را میزنیم یک عملیات 
    shuffling انجام میشود
    در این قسمت از شما انتظار می‌رود کوئری جوین زیر را نوشته و اجرا کنید همچنین به  صفحه 
    application master ui
    مراجعه کنید و نحوه shuffleing را گزارش کنید . 
     و همچنین توضیح دهید DAG scheduler  در اسپارک چیست ؟

یکی از عملیات های اولیه در برنامه نویسی حافظه در آپاچی اسپارک، شافلینگ است. این فرآیند مسئول مدیریت توزیع مجدد داده ها در طول کارهای Spark است. هنگامی که یک کار Spark نیاز به جابجایی داده ها بین اجراکننده ها یا بین executors و driver دارد، سرویس shuffle این حرکت داده را مدیریت می کند.
برای سازمان‌دهی داده‌ها، Spark shuffle مجموعه‌ای از وظایف را تولید می‌کند – وظایف نقشه‌ای برای سازماندهی داده‌ها و مجموعه‌ای از وظایف کاهش‌یافته برای جمع‌آوری آن‌ها.
برای دانستن اینکه چه زمانی یک تصادف رخ می دهد، مهم است بداینم که چگونه زمان بندی در Spark برای بارهای کاری روی خوشه کار می کند. به طور کلی، این مخلوط بین دو مرحله اتفاق می افتد. هنگامی که DAGScheduler اجرای فیزیکی طرح را برای طرح های منطقی تولید می کند، تمام RDD ها را به یک مرحله واحد متصل می کند که از طریق وابستگی باریک به هم متصل شده اند.
در نتیجه، طرح منطقی بین RDD ها با وابستگی های گسترده را قطع خواهد کرد. آخرین مرحله ای که تولید می کند به عنوان ResultStage شناخته می شود. همه انواع مراحل برای محاسبه ResultStage، معروف به ShuffleMapStages، مورد نیاز است.
   . سرویس Shuffle هم می تواند بر روی یک گره Spark Worker  حتی خارج از Spark Worker اجرا شود. Spark Executor می تواند در آن اجرا شود. در طول فرآیند ثبت نام هم مجری سرویس را در مورد مکانی روی دیسک که در آن فایل های shuffle را ذخیره کرده مطلع می سازد. سرویس Shuffle فایل‌های Shuffle را تماشا کرده و اگر درخواستی دیگر از بود فایل‌های shuffle را بر خواهد گرداند.

# تابع تعریف شده توسط کاربر(UDF)

    یکی از مزایای اسپارک این است که نه تنها برای ما یک زبان SQL فراهم کرده که روی چندین سرور به صورت همزمان پردازش را انجام دهد بلکه 
    میتوان به زبان‌های مختلف توابعی تعریف کرد که روی  همه executor ها اجرا شود
    در این بخش از شما انتظار می‌رود که با نوشتن یک UDF به زبان پایتونی 
    توضیحات هر اسلحه را به 3 کلمه اول آن کوتاه کنید و بقیه کلمات را حذف کنید و سوتونی به نام short_description را به داده های 
    Weapon اضافه کنید
    
    
 </div>
 
[pyspark.sql.functions.udf](https://spark.apache.org/docs/3.1.3/api/python/reference/api/pyspark.sql.functions.udf.html)

[pyspark.sql.DataFrame.withColumn](https://spark.apache.org/docs/3.1.3/api/python/reference/api/pyspark.sql.DataFrame.withColumn.html)

    



```python
from pyspark.sql.types import StringType

@udf(returnType=StringType()) 
def short_description_function(string):
    return ' '.join(string.split()[:3])

Weapon.withColumn("short_description", short_description_function(Weapon.Description)).show()
```
